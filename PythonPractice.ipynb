```
## Class to define new types
class Point:
    def __init__(self, x, y):  ## This is constructor of class
        self.x = x
        self.y = y
    def move(self):
        print("move")
## Creating object from class blueprints or template
point1 = Point(5,2)
point2 = Point(6,4)

## Inheritance  one class get its properties/attributes from another class through inheritance.
class Mammal:
    def walk(self):
        print("walk")
class Dog(Mammal):
    def bark(self):
        print("bark")

dog = Dog() 
dog.walk()  ## inherited from Mammal
dog.bark()  ## defined in Dog

```

```
## filter, map, reduce
numbers = ["3", '34', '64']
def sq(a):
    return a*a
def is_greater(a):
    return a > 30
num = list(map(lambda x: int(x), numbers))
print(list(num))
print(list(map(sq,num)))

gr_than_30 = list(filter(is_greater, num))
gr_than_35 = list(filter(lambda x: x>35, num))
print(gr_than_35)

from functools import reduce
sum = reduce(lambda x, y: x + y, num, 0)
print(sum)
```
